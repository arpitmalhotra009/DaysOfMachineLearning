```python
#importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
plt.style.use("ggplot")
%matplotlib inline
```

```python
#importing the data and getting the information of the dataset
data = pd.read_csv("DMV_Written_Tests.csv")
print(data.head()) # Printing top 5 rows
print("\n\n")
print(data.info()) #Printing data information 
```

## Data Visualization

```python
fig, ax = plt.subplots(figsize=(10,8))
ax = sns.scatterplot(x = data.loc[data['Results']==1, 'DMV_Test_1'],
                     y = data.loc[data['Results']==1, 'DMV_Test_2'],
                     marker = "^",
                     color = "green",
                     s = 60)
sns.scatterplot(x = data.loc[data['Results']==0, 'DMV_Test_1'],
                y = data.loc[data['Results']==0, 'DMV_Test_2'],
                marker = "X",
                color = "red",
                s = 60)

ax.set(xlabel="DMV Written Test 1 Scores", ylabel="DMV Written Test 2 Scores")
ax.legend(["Passed", "Failed"])
plt.show()
```

**It can be obesrved from the above plotted graph that the passed or persons who have got the license lie on the upper right hand side of the graph and the persons who have failed to get the Driver's License lie on the lower left of the graph. By closely observing the data we can say that a line can be drawn on the graph which can help us in predicting whether a person will get the license or not based on the scores in DMV written tests**
